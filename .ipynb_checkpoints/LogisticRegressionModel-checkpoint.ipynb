```python
import pandas as pd#read in the data using pandas
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("data/data without infertility _final.csv")#check data has been read in properly
df.head()
```

```python
#create a dataframe with all training data except the target column
X = df.drop(columns=["PCOS (Y/N)"])#check that the target variable has been removed
X.head()
```

```python
#separate target values
y = df["PCOS (Y/N)"].values#view target values
y[0:5]
```

```python
#split dataset into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)# Fit the classifier to the data
knn.fit(X_train,y_train)
```

```python
#show first 5 model predictions on the test data
knn.predict(X_test)[0:5]
```

```python
#check accuracy of our model on the test data
knn.score(X_test, y_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/data without infertility _final.csv")
del df["Unnamed: 42"]
del df["Sl. No"]
del df["Patient File No."]
del df["FSH/LH"]
del df["Waist:Hip Ratio"]

df = df.drop(df.index[[200,161,223,296,305,216,195,191,329,455]])

null_columns = df.columns[df.isnull().any()]
df[null_columns].isnull().sum()

df = df.dropna()
```

```python
for col in df.columns:
    print(col,"\nmin : ",min(df[col]),", max:",max(df[col]),"\n")
```

```python
for col in df.columns:    
    if(col.strip()[-5:]) != "(Y/N)":
        plt.scatter([var for var in range(len(df[col]))], df[col])
        plt.xlabel('Sr. No.')
        plt.ylabel(col)
        plt.show()
```
